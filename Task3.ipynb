### q16 : Write a fun which will take input as a dict and give me out as a list of all the values even in case of 2 level nesting it should work . 


```python
def dict_list(d):
    l = []
    for i in d.items():
        if type(i) != dict:
            for j in i:
                if type(j) != dict:
                    l.append(j)
    return l
        
```

```python
x = {'k1':'v1', 'k2':'v2'}
dict_list(x)
```

### q17 : write a function which will take multiple list as a input and give me concatnation of all the element as and output


```python
def list_conc(*arg):
    l = []
    for i in arg:
        for j in i:
            for k in j:
                l.append(k)
        
    return l
```

```python
list_conc([[56,23], [45,76]])
```

### q22 : write a function which can read video file and play for you . 


```python
import cv2
import numpy as np

# create a video capture object and read from input file
# if the input is the camera, pass 0 instead of the video file name
cap = cv2.VideoCapture('video.mp4')

# Read until video is completed
while (cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True:
        
        #display the resulting frame
        cv2.imshow('Frame', frame)
        
        # press Q on keyboard to exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
            
    else:
        break
#when everything done, release the video capture object
cap.release()

#closes all the frames
cv2.destroyAllWindows()
```
